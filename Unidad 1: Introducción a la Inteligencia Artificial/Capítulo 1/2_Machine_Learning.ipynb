# **1. ¿QUÉ ES MACHINE LEARNING?**

El **Machine Learning (ML)** o **aprendizaje automático** es un subconjunto de la inteligencia artificial que permite que las computadoras aprendan a partir de datos, mejorando su rendimiento con la experiencia sin estar programadas de forma explícita para cada tarea.

---

> **Ejemplo:**

- Si programas una máquina para que reconozca gatos, no le das todas las reglas posibles ("tienen 4 patas, orejas puntiagudas, bigotes...").

- - En lugar de eso, le das miles de fotos de gatos y no-gatos.

- - El algoritmo aprende patrones (formas, colores, texturas) y con eso puede predecir si una nueva imagen tiene un gato o no.

---

> **¿Cómo funciona?**

1. Se le da un conjunto de datos (ej: imágenes, texto, números).

2. El algoritmo busca patrones en esos datos.

3. Hace predicciones sobre nuevos casos basados en lo aprendido.

4. Se ajusta cuando comete errores para mejorar su precisión.

---

> **Tipos de Machine Learning:**

1. Aprendizaje supervisado → se entrena con datos etiquetados (ejemplo: imágenes con la etiqueta “perro” o “gato”).

2. Aprendizaje no supervisado → busca patrones sin etiquetas (ejemplo: agrupar clientes según hábitos de compra).

3. Aprendizaje por refuerzo → aprende por prueba y error recibiendo recompensas o castigos (ejemplo: un robot aprendiendo a caminar).

---

> **Ejemplos de ML en la vida real:**

- Netflix o Spotify recomendándote series y música.

- Filtros de spam en Gmail.

- Coches autónomos aprendiendo a conducir.

- Reconocimiento de voz en asistentes como Siri o Alexa.

---

# **3. APRENDIZAJE AUTOMÁTICO VS ESTADÍSTICA**

El **aprendizaje automático (Machine Learning, ML)** y la **estadística** están estrechamente relacionados porque ambos analizan datos para obtener conclusiones. Sin embargo, **tienen enfoques distintos**.

---

## **Enfoque de la Estadística**

* Busca **entender los datos** y explicar las relaciones entre variables.
* Se centra en **modelos interpretables** (fáciles de entender por humanos).
* Objetivo: **explicar y validar hipótesis**.
* Ejemplo: analizar si el consumo de café está relacionado con mejor rendimiento académico.

---

## **Enfoque del Machine Learning**

* Busca **hacer predicciones automáticas y precisas**.
* Se centra en **modelos que funcionan bien, aunque sean complejos** (ej: redes neuronales).
* Objetivo: **predecir y generalizar a nuevos datos**.
* Ejemplo: entrenar un modelo que prediga la probabilidad de que un estudiante apruebe un curso según sus hábitos de estudio.

---

## Diferencias clave

| Aspecto            | Estadística                                     | Machine Learning                                               |
| ------------------ | ----------------------------------------------- | -------------------------------------------------------------- |
| **Meta**           | Explicar y validar relaciones                   | Predecir resultados con precisión                              |
| **Enfoque**        | Basado en teorías y modelos matemáticos         | Basado en algoritmos y datos masivos                           |
| **Datos**          | Funciona bien con conjuntos pequeños            | Requiere grandes volúmenes de datos                            |
| **Interpretación** | Modelos interpretables (regresión, ANOVA, etc.) | Modelos a veces “caja negra” (deep learning)                   |
| **Uso típico**     | Investigación científica, encuestas, economía   | Reconocimiento de voz, visión por computadora, recomendaciones |

---

## Ejemplo comparativo

Supongamos que queremos estudiar la **relación entre el peso y la altura de las personas**:

* **Estadística**: usa una **regresión lineal** para explicar cómo cambia el peso según la altura, con una ecuación clara:
  $peso = a + b \times altura$
  (puedes interpretar qué significa cada parámetro).

* **Machine Learning**: puede usar **árboles de decisión, random forest o redes neuronales** para predecir el peso con mucha precisión, incluso detectando relaciones no lineales, pero el modelo podría ser difícil de interpretar.

---

> En resumen:

* **Estadística** = explicar y entender relaciones.
* **Machine Learning** = predecir con precisión y automatizar.

# **4. PRINCIPALES TAREAS DE LA MINERÍA DE DATOS**

Al analizar una gran cantidad de datos con tecnología de aprendizaje automático, se pueden encontrar patrones que eran invisibles externamente, lo que se denomina minería de datos. La minería de datos se ocupa de los siguientes cuatro problemas:

---

### **Predicción**

* Consiste en usar datos históricos para **predecir valores futuros**.
* Ejemplo: predecir el precio de una acción para la próxima semana a partir de datos pasados.
* Herramientas típicas: regresión, series de tiempo, redes neuronales.

---

### **Agrupamiento (Clustering)**

* Se trata de **agrupar objetos similares** en categorías sin que estén previamente etiquetados.
* Ejemplo: segmentar clientes de un supermercado en grupos según sus hábitos de compra (sin que alguien les haya puesto etiquetas antes).
* Algoritmos típicos: K-Means, DBSCAN, clustering jerárquico.

---

### **Clasificación**

* Es asignar cada dato a una **categoría ya conocida** (etiquetas predefinidas).
* Ejemplo: clasificar correos electrónicos en **spam** o **no spam**.
* Algoritmos típicos: árboles de decisión, SVM, Naive Bayes.

---

### **Regla de asociación**

* Busca **relaciones o correlaciones** entre variables dentro de grandes bases de datos.
* Ejemplo: descubrir que los clientes que compran **leche** también suelen comprar **pan** (reglas tipo *“si X, entonces Y”*).
* Algoritmo típico: Apriori, FP-Growth.

---

**Resumen visual**:

| Tarea                   | Qué hace                         | Ejemplo                              |
| ----------------------- | -------------------------------- | ------------------------------------ |
| **Predicción**          | Anticipa valores futuros         | Precio de acciones                   |
| **Agrupamiento**        | Agrupa datos similares           | Segmentación de clientes             |
| **Clasificación**       | Asigna datos a clases conocidas  | Detección de spam                    |
| **Regla de asociación** | Encuentra relaciones entre ítems | “Si compra pañales → compra cerveza” |

# **5. APRENDIZAJE SUPERVISADO**

El **aprendizaje supervisado** es un tipo de **aprendizaje automático (Machine Learning)** en el que el modelo se entrena con un conjunto de datos **etiquetados**, es decir, datos de entrada junto con la respuesta correcta (*output*).

La idea es que la máquina **aprenda a relacionar las entradas con las salidas** y luego pueda predecir la salida de nuevos datos que nunca ha visto.

---

## Características clave:

1. **Datos etiquetados**: cada ejemplo del entrenamiento incluye la respuesta correcta.

   * Ejemplo: una foto (entrada) con la etiqueta “perro” o “gato” (salida).
2. **Objetivo**: que el modelo aprenda una **función de predicción**.
3. **Aplicación**: se usa para **clasificación** (categorías) y **regresión** (valores numéricos).

---

## Ejemplos:

* **Clasificación**:

  * Entrada → email
  * Salida → “spam” o “no spam”.

* **Regresión**:

  * Entrada → tamaño y ubicación de una casa
  * Salida → precio estimado de la casa.

---

## Pasos del aprendizaje supervisado:

1. Recolectar datos etiquetados.
2. Dividir los datos en **entrenamiento** y **prueba**.
3. Entrenar el modelo con el conjunto de entrenamiento.
4. Evaluarlo con datos de prueba.
5. Usarlo para predecir nuevos casos.

---

> **Resumen sencillo**:

El **aprendizaje supervisado** funciona como un profesor enseñando a un estudiante: le muestra ejemplos con la respuesta correcta, el estudiante aprende la relación, y después puede responder correctamente cuando le pongan un examen con ejemplos nuevos.

---


## **Principales Ténicas y Algoritmos del Aprendizaje Supervisado**

El **aprendizaje supervisado** tiene varias técnicas y algoritmos que se usan según el tipo de problema (clasificación o regresión).

---

### 1. **Regresión Lineal**

* Se usa para **regresión** (predecir valores numéricos).
* Encuentra una línea recta (o hiperplano) que mejor se ajuste a los datos.
* Ejemplo: predecir el **precio de una casa** según su tamaño.

---

### 2. **Regresión Logística**

* Se usa para **clasificación binaria** (dos clases).
* Predice la probabilidad de pertenecer a una clase.
* Ejemplo: clasificar un correo en **spam** o **no spam**.

---

### 3. **K-Nearest Neighbors (KNN)**

* Clasifica un dato nuevo observando a sus **vecinos más cercanos**.
* Funciona tanto para clasificación como regresión.
* Ejemplo: recomendar películas según las que vieron usuarios “similares”.

---

### 4. **Árboles de Decisión**

* Divide los datos en **ramas con condiciones** hasta llegar a una decisión.
* Muy interpretables.
* Ejemplo: decidir si dar o no un préstamo según ingresos, edad y deudas.

---

### 5. **Random Forest**

* Conjunto de **muchos árboles de decisión** combinados para mejorar precisión.
* Muy usado porque reduce errores y sobreajuste.
* Ejemplo: detección de fraudes en tarjetas de crédito.

---

### 6. **Support Vector Machines (SVM)**

* Encuentra el **hiperplano óptimo** que separa las clases con el mayor margen posible.
* Funciona muy bien en problemas de clasificación compleja.
* Ejemplo: reconocer escritura manuscrita (números o letras).

---

### 7. **Naive Bayes**

* Algoritmo probabilístico basado en el **Teorema de Bayes**.
* Rápido y funciona bien en clasificación de texto.
* Ejemplo: clasificación de correos electrónicos (spam/no spam).

---

### 8. **Redes Neuronales Artificiales**

* Inspiradas en el cerebro humano.
* Usadas tanto en clasificación como regresión.
* Ejemplo: reconocimiento de imágenes (perros, gatos, rostros).

---

📌 **Resumen práctico**:

| Algoritmo               | Tipo principal            | Ejemplo                     |
| ----------------------- | ------------------------- | --------------------------- |
| **Regresión lineal**    | Regresión                 | Precio de casas             |
| **Regresión logística** | Clasificación             | Spam / No spam              |
| **KNN**                 | Clasificación / Regresión | Recomendación de películas  |
| **Árboles de decisión** | Ambos                     | Aprobado / Reprobado        |
| **Random Forest**       | Ambos                     | Detección de fraudes        |
| **SVM**                 | Clasificación             | Reconocimiento de escritura |
| **Naive Bayes**         | Clasificación             | Clasificación de texto      |
| **Redes Neuronales**    | Ambos                     | Reconocimiento de imágenes  |

# **6. APRENDIZAJE NO SUPERVISADO**

El **aprendizaje no supervisado** es un tipo de **aprendizaje automático (Machine Learning)** en el que el modelo trabaja con **datos no etiquetados**.

A diferencia del aprendizaje supervisado, aquí no existen respuestas correctas predefinidas; la máquina debe **encontrar patrones ocultos, estructuras o relaciones en los datos por sí misma**.

---

## Características principales

1. **No hay etiquetas**: el algoritmo no sabe cuál es la “respuesta correcta”.
2. **Objetivo**: descubrir estructura en los datos (agrupamientos, asociaciones, reducción de dimensiones).
3. **Se usa mucho para explorar datos** cuando no conocemos bien su distribución.

---

## Técnicas y algoritmos comunes

1. **Clustering (Agrupamiento)**

   * Agrupa datos similares en “clusters”.
   * Ejemplo: segmentar clientes de un banco en grupos según sus hábitos financieros.
   * Algoritmos: K-Means, DBSCAN, Clustering jerárquico (HCA), Detección de Anomalías y Detección de Valores Atíícos, SVM de una clase, Bosque de aislamiento.

2. **Reglas de Asociación**

   * Busca relaciones entre variables en grandes conjuntos de datos.
   * Ejemplo: descubrir que quienes compran **pan** también suelen comprar **mantequilla**.
   * Algoritmo: Apriori, FP-Growth, Eclat.

3. **Reducción de Dimensionalidad**

   * Reduce la cantidad de variables manteniendo la información esencial.
   * Ejemplo: compresión de imágenes.
   * Algoritmos: PCA (Análisis de Componentes Principales), t-SNE, Kernel PCA, Incrustación lineal local.

---

## Ejemplos de aplicación

* Segmentación de mercado.
* Recomendadores de productos (Amazon, Netflix).
* Detección de anomalías (fraude en tarjetas de crédito, fallos en sensores).
* Agrupamiento de documentos o noticias similares.

---

**Resumen sencillo**:

El **aprendizaje supervisado** es como un profesor que te enseña con ejemplos y respuestas correctas.
El **aprendizaje no supervisado** es como un explorador que tiene un montón de datos sin etiquetas y debe **descubrir por sí mismo** cómo se organizan.

---


## **Principales Técnicas o Algoritmos del Aprendizaje No Supervisado**

### 1. **Clustering (Agrupamiento)**

Consiste en **descubrir patrones ocultos** al agrupar datos en “clusters” o grupos sin que nadie le diga al modelo cuáles deberían ser esos grupos.

* **¿Para qué sirve?**

  * Segmentación de clientes (marketing, bancos, e-commerce).
  * Detección de anomalías (fraude bancario, fallas en sensores).
  * Biología (agrupar genes o células con funciones similares).
  * Sistemas de recomendación (agrupar usuarios con gustos parecidos).

* **Principales algoritmos:**

  * **K-Means** → Divide los datos en *k* grupos minimizando la distancia al centroide. Es rápido pero sensible a valores atípicos.
  * **DBSCAN** → Detecta clusters de distintas formas/tamaños, además detecta *outliers*. Muy usado en datos geoespaciales.
  * **Clustering Jerárquico (HCA)** → Construye una jerarquía de grupos (tipo árbol). Bueno para análisis exploratorio.
  * **SVM de una clase (One-Class SVM)** → Especializado en detectar si un dato es “normal” o una anomalía.
  * **Isolation Forest** → Aísla puntos anómalos en datasets grandes (muy usado en ciberseguridad y fraude).

---

### 2. **Reglas de Asociación**

Aquí el objetivo es **encontrar relaciones frecuentes entre ítems o eventos** en un conjunto de datos.

* **¿Para qué sirve?**

  * *Market Basket Analysis* (ejemplo clásico: si compras pan, también compras mantequilla).
  * Recomendadores de productos (Amazon, Mercado Libre).
  * Detección de patrones en bioinformática, redes sociales, seguridad.

* **Principales algoritmos:**

  * **Apriori** → Encuentra reglas frecuentes combinando ítems paso a paso. Fácil de entender, pero pesado con muchos datos.
  * **FP-Growth (Frequent Pattern Growth)** → Optimización de Apriori que construye una estructura tipo árbol para mejorar velocidad.
  * **Eclat** → Usa intersección de conjuntos en vez de combinaciones, suele ser más rápido en algunos escenarios.

---

### 3. **Reducción de Dimensionalidad**

Cuando los datos tienen muchísimas variables (imágenes, texto, genómica, big data), se busca **reducir dimensiones** pero conservando la mayor información posible.

* **¿Para qué sirve?**

  * Preprocesar datos antes de aplicar clustering o clasificación.
  * Visualización de datos en 2D/3D.
  * Compresión de imágenes y reducción de ruido.
  * Evitar el problema de la “maldición de la dimensionalidad” (cuando demasiadas variables confunden al modelo).

* **Principales algoritmos:**

  * **PCA (Principal Component Analysis)** → Proyecta los datos en nuevas variables (componentes principales) que maximizan la varianza.
  * **Kernel PCA** → Variante de PCA que usa kernels para capturar relaciones no lineales.
  * **t-SNE (t-Distributed Stochastic Neighbor Embedding)** → Excelente para visualización en 2D/3D de datos muy complejos. Muy usado en biología computacional y NLP.
  * **LLE (Locally Linear Embedding)** → Preserva la estructura local de los datos, útil para datos no lineales.

---

En resumen:

* **Clustering** = agrupar sin etiquetas.
* **Reglas de Asociación** = descubrir patrones de co-ocurrencia.
* **Reducción de Dimensionalidad** = simplificar datos sin perder demasiada información.

# 7. **ALGORITMOS DE APRENDIZAJE AUTOMÁTICO**



**formas de entrenar y aplicar algoritmos de aprendizaje automático**.

---

## 1. **Aprendizaje por Lotes (Batch Learning)**

* El modelo se entrena con **todo el conjunto de datos de una sola vez**.
* Cada vez que llegan nuevos datos, hay que **volver a entrenar desde cero** con el dataset completo.

> **Ejemplo:** entrenar un modelo de predicción de fraudes con todos los registros de transacciones bancarias históricas. Si entran nuevos clientes, se debe reentrenar el modelo completo.

✅ Ventajas: mayor precisión en modelos grandes.

❌ Desventajas: consume muchos recursos y no se adapta rápido a cambios.

---

## 2. **Aprendizaje en Línea (Online Learning)**

* El modelo **aprende de manera incremental**: se actualiza cada vez que llegan nuevos datos, sin necesidad de reentrenar todo.
* Ideal para sistemas donde los datos cambian rápido o llegan en flujo continuo (*streaming*).

> **Ejemplo:** un sistema de recomendaciones de YouTube/Netflix que va aprendiendo de lo que ves en tiempo real y adapta las sugerencias.

✅ Ventajas: se adapta rápido a nuevos datos, eficiente en memoria.

❌ Desventajas: puede ser inestable si los datos que llegan son muy ruidosos.

---

## 3. **Aprendizaje Basado en Instancias (Instance-Based Learning)**

* El modelo **no construye reglas generales al entrenar**, sino que **guarda ejemplos del dataset** y decide comparando con ellos.
* Aprende “memorizar y comparar” más que “generalizar”.

> **Ejemplo:** el algoritmo **K-Nearest Neighbors (KNN)**. Para clasificar una nueva fruta, compara sus características (peso, color, tamaño) con las frutas almacenadas y decide según las más cercanas.

✅ Ventajas: simple y preciso si los datos están bien representados.
❌ Desventajas: lento con grandes volúmenes de datos y sensible a ruido.

---

## 4. **Aprendizaje Basado en Modelos (Model-Based Learning)**

* El modelo **aprende una función matemática general** a partir de los datos de entrenamiento.
* No guarda todos los ejemplos, sino que extrae reglas/patrones y los usa para predecir.

> **Ejemplo:** la **regresión lineal** para predecir precios de casas: el modelo aprende la fórmula `Precio = a*metros² + b*habitaciones + c`.

> Otro ejemplo: una red neuronal que aprende pesos y sesgos para predecir imágenes de gatos.

✅ Ventajas: generaliza bien y es eficiente en predicciones.

❌ Desventajas: si el modelo es muy simple, puede no capturar toda la complejidad de los datos.

---

En resumen:

* **Batch** = aprende todo de golpe.
* **Online** = aprende poco a poco con nuevos datos.
* **Instance-based** = memoriza ejemplos y compara.
* **Model-based** = aprende una regla general y predice con ella.

---
