# **1. ¬øQU√â ES MACHINE LEARNING?**

# **3. APRENDIZAJE AUTOM√ÅTICO VS ESTAD√çSTICA**

El **aprendizaje autom√°tico (Machine Learning, ML)** y la **estad√≠stica** est√°n estrechamente relacionados porque ambos analizan datos para obtener conclusiones. Sin embargo, **tienen enfoques distintos**.

## **Enfoque de la Estad√≠stica**

* Busca **entender los datos** y explicar las relaciones entre variables.
* Se centra en **modelos interpretables** (f√°ciles de entender por humanos).
* Objetivo: **explicar y validar hip√≥tesis**.
* Ejemplo: analizar si el consumo de caf√© est√° relacionado con mejor rendimiento acad√©mico.

## **Enfoque del Machine Learning**

* Busca **hacer predicciones autom√°ticas y precisas**.
* Se centra en **modelos que funcionan bien, aunque sean complejos** (ej: redes neuronales).
* Objetivo: **predecir y generalizar a nuevos datos**.
* Ejemplo: entrenar un modelo que prediga la probabilidad de que un estudiante apruebe un curso seg√∫n sus h√°bitos de estudio.

## Diferencias clave

| Aspecto            | Estad√≠stica                                     | Machine Learning                                               |
| ------------------ | ----------------------------------------------- | -------------------------------------------------------------- |
| **Meta**           | Explicar y validar relaciones                   | Predecir resultados con precisi√≥n                              |
| **Enfoque**        | Basado en teor√≠as y modelos matem√°ticos         | Basado en algoritmos y datos masivos                           |
| **Datos**          | Funciona bien con conjuntos peque√±os            | Requiere grandes vol√∫menes de datos                            |
| **Interpretaci√≥n** | Modelos interpretables (regresi√≥n, ANOVA, etc.) | Modelos a veces ‚Äúcaja negra‚Äù (deep learning)                   |
| **Uso t√≠pico**     | Investigaci√≥n cient√≠fica, encuestas, econom√≠a   | Reconocimiento de voz, visi√≥n por computadora, recomendaciones |

---

## Ejemplo comparativo

Supongamos que queremos estudiar la **relaci√≥n entre el peso y la altura de las personas**:

* **Estad√≠stica**: usa una **regresi√≥n lineal** para explicar c√≥mo cambia el peso seg√∫n la altura, con una ecuaci√≥n clara:
  $peso = a + b \times altura$

* **Machine Learning**: puede usar **√°rboles de decisi√≥n, random forest o redes neuronales** para predecir el peso con mucha precisi√≥n, incluso detectando relaciones no lineales, pero el modelo podr√≠a ser dif√≠cil de interpretar.

> En resumen:

* **Estad√≠stica** = explicar y entender relaciones.

# **4. PRINCIPALES TAREAS DE LA MINER√çA DE DATOS**

Al analizar una gran cantidad de datos con tecnolog√≠a de aprendizaje autom√°tico, se pueden encontrar patrones que eran invisibles externamente, lo que se denomina miner√≠a de datos. La miner√≠a de datos se ocupa de los siguientes cuatro problemas:

---

### **Predicci√≥n**

* Consiste en usar datos hist√≥ricos para **predecir valores futuros**.
* Ejemplo: predecir el precio de una acci√≥n para la pr√≥xima semana a partir de datos pasados.
* Herramientas t√≠picas: regresi√≥n, series de tiempo, redes neuronales.

### **Agrupamiento (Clustering)**

* Se trata de **agrupar objetos similares** en categor√≠as sin que est√©n previamente etiquetados.
* Ejemplo: segmentar clientes de un supermercado en grupos seg√∫n sus h√°bitos de compra (sin que alguien les haya puesto etiquetas antes).

### **Clasificaci√≥n**

* Es asignar cada dato a una **categor√≠a ya conocida** (etiquetas predefinidas).
* Ejemplo: clasificar correos electr√≥nicos en **spam** o **no spam**.
* Algoritmos t√≠picos: √°rboles de decisi√≥n, SVM, Naive Bayes.

### **Regla de asociaci√≥n**

* Busca **relaciones o correlaciones** entre variables dentro de grandes bases de datos.
* Ejemplo: descubrir que los clientes que compran **leche** tambi√©n suelen comprar **pan** (reglas tipo *‚Äúsi X, entonces Y‚Äù*).
* Algoritmo t√≠pico: Apriori, FP-Growth.

---

# **5. APRENDIZAJE SUPERVISADO**

El **aprendizaje supervisado** es un tipo de **aprendizaje autom√°tico (Machine Learning)** en el que el modelo se entrena con un conjunto de datos **etiquetados**, es decir, datos de entrada junto con la respuesta correcta (*output*).

La idea es que la m√°quina **aprenda a relacionar las entradas con las salidas** y luego pueda predecir la salida de nuevos datos que nunca ha visto.

---

## Caracter√≠sticas clave:

1. **Datos etiquetados**: cada ejemplo del entrenamiento incluye la respuesta correcta.

   * Ejemplo: una foto (entrada) con la etiqueta ‚Äúperro‚Äù o ‚Äúgato‚Äù (salida).
2. **Objetivo**: que el modelo aprenda una **funci√≥n de predicci√≥n**.
3. **Aplicaci√≥n**: se usa para **clasificaci√≥n** (categor√≠as) y **regresi√≥n** (valores num√©ricos).

---

## Ejemplos:

* **Clasificaci√≥n**:

  * Entrada ‚Üí email
  * Salida ‚Üí ‚Äúspam‚Äù o ‚Äúno spam‚Äù.

* **Regresi√≥n**:

  * Entrada ‚Üí tama√±o y ubicaci√≥n de una casa
  * Salida ‚Üí precio estimado de la casa.

---

## Pasos del aprendizaje supervisado:

1. Recolectar datos etiquetados.
2. Dividir los datos en **entrenamiento** y **prueba**.
3. Entrenar el modelo con el conjunto de entrenamiento.
4. Evaluarlo con datos de prueba.
5. Usarlo para predecir nuevos casos.

---

> **Resumen sencillo**:

El **aprendizaje supervisado** funciona como un profesor ense√±ando a un estudiante: le muestra ejemplos con la respuesta correcta, el estudiante aprende la relaci√≥n, y despu√©s puede responder correctamente cuando le pongan un examen con ejemplos nuevos.

---


## **Principales T√©nicas y Algoritmos del Aprendizaje Supervisado**

El **aprendizaje supervisado** tiene varias t√©cnicas y algoritmos que se usan seg√∫n el tipo de problema (clasificaci√≥n o regresi√≥n).

---

### 1. **Regresi√≥n Lineal**

* Se usa para **regresi√≥n** (predecir valores num√©ricos).
* Encuentra una l√≠nea recta (o hiperplano) que mejor se ajuste a los datos.
* Ejemplo: predecir el **precio de una casa** seg√∫n su tama√±o.

---

### 2. **Regresi√≥n Log√≠stica**

* Se usa para **clasificaci√≥n binaria** (dos clases).
* Predice la probabilidad de pertenecer a una clase.
* Ejemplo: clasificar un correo en **spam** o **no spam**.

---

### 3. **K-Nearest Neighbors (KNN)**

* Clasifica un dato nuevo observando a sus **vecinos m√°s cercanos**.
* Funciona tanto para clasificaci√≥n como regresi√≥n.
* Ejemplo: recomendar pel√≠culas seg√∫n las que vieron usuarios ‚Äúsimilares‚Äù.

---

### 4. **√Årboles de Decisi√≥n**

* Divide los datos en **ramas con condiciones** hasta llegar a una decisi√≥n.
* Muy interpretables.
* Ejemplo: decidir si dar o no un pr√©stamo seg√∫n ingresos, edad y deudas.

---

### 5. **Random Forest**

* Conjunto de **muchos √°rboles de decisi√≥n** combinados para mejorar precisi√≥n.
* Muy usado porque reduce errores y sobreajuste.
* Ejemplo: detecci√≥n de fraudes en tarjetas de cr√©dito.

---

### 6. **Support Vector Machines (SVM)**

* Encuentra el **hiperplano √≥ptimo** que separa las clases con el mayor margen posible.
* Funciona muy bien en problemas de clasificaci√≥n compleja.
* Ejemplo: reconocer escritura manuscrita (n√∫meros o letras).

---

### 7. **Naive Bayes**

* Algoritmo probabil√≠stico basado en el **Teorema de Bayes**.
* R√°pido y funciona bien en clasificaci√≥n de texto.
* Ejemplo: clasificaci√≥n de correos electr√≥nicos (spam/no spam).

---

### 8. **Redes Neuronales Artificiales**

* Inspiradas en el cerebro humano.
* Usadas tanto en clasificaci√≥n como regresi√≥n.
* Ejemplo: reconocimiento de im√°genes (perros, gatos, rostros).

---

üìå **Resumen pr√°ctico**:

| Algoritmo               | Tipo principal            | Ejemplo                     |
| ----------------------- | ------------------------- | --------------------------- |
| **Regresi√≥n lineal**    | Regresi√≥n                 | Precio de casas             |
| **Regresi√≥n log√≠stica** | Clasificaci√≥n             | Spam / No spam              |
| **KNN**                 | Clasificaci√≥n / Regresi√≥n | Recomendaci√≥n de pel√≠culas  |
| **√Årboles de decisi√≥n** | Ambos                     | Aprobado / Reprobado        |
| **Random Forest**       | Ambos                     | Detecci√≥n de fraudes        |
| **SVM**                 | Clasificaci√≥n             | Reconocimiento de escritura |
| **Naive Bayes**         | Clasificaci√≥n             | Clasificaci√≥n de texto      |
| **Redes Neuronales**    | Ambos                     | Reconocimiento de im√°genes  |

# **6. APRENDIZAJE NO SUPERVISADO**

El **aprendizaje no supervisado** es un tipo de **aprendizaje autom√°tico (Machine Learning)** en el que el modelo trabaja con **datos no etiquetados**.

A diferencia del aprendizaje supervisado, aqu√≠ no existen respuestas correctas predefinidas; la m√°quina debe **encontrar patrones ocultos, estructuras o relaciones en los datos por s√≠ misma**.

---

## Caracter√≠sticas principales

1. **No hay etiquetas**: el algoritmo no sabe cu√°l es la ‚Äúrespuesta correcta‚Äù.
2. **Objetivo**: descubrir estructura en los datos (agrupamientos, asociaciones, reducci√≥n de dimensiones).
3. **Se usa mucho para explorar datos** cuando no conocemos bien su distribuci√≥n.

---

## T√©cnicas y algoritmos comunes

1. **Clustering (Agrupamiento)**

   * Agrupa datos similares en ‚Äúclusters‚Äù.
   * Ejemplo: segmentar clientes de un banco en grupos seg√∫n sus h√°bitos financieros.
   * Algoritmos: K-Means, DBSCAN, Clustering jer√°rquico (HCA), Detecci√≥n de Anomal√≠as y Detecci√≥n de Valores At√≠√≠cos, SVM de una clase, Bosque de aislamiento.

2. **Reglas de Asociaci√≥n**

   * Busca relaciones entre variables en grandes conjuntos de datos.
   * Ejemplo: descubrir que quienes compran **pan** tambi√©n suelen comprar **mantequilla**.
   * Algoritmo: Apriori, FP-Growth, Eclat.

3. **Reducci√≥n de Dimensionalidad**

   * Reduce la cantidad de variables manteniendo la informaci√≥n esencial.
   * Ejemplo: compresi√≥n de im√°genes.
   * Algoritmos: PCA (An√°lisis de Componentes Principales), t-SNE, Kernel PCA, Incrustaci√≥n lineal local.

---

## Ejemplos de aplicaci√≥n

* Segmentaci√≥n de mercado.
* Recomendadores de productos (Amazon, Netflix).
* Detecci√≥n de anomal√≠as (fraude en tarjetas de cr√©dito, fallos en sensores).
* Agrupamiento de documentos o noticias similares.

---

**Resumen sencillo**:

El **aprendizaje supervisado** es como un profesor que te ense√±a con ejemplos y respuestas correctas.
El **aprendizaje no supervisado** es como un explorador que tiene un mont√≥n de datos sin etiquetas y debe **descubrir por s√≠ mismo** c√≥mo se organizan.

---


## **Principales T√©cnicas o Algoritmos del Aprendizaje No Supervisado**

### 1. **Clustering (Agrupamiento)**

Consiste en **descubrir patrones ocultos** al agrupar datos en ‚Äúclusters‚Äù o grupos sin que nadie le diga al modelo cu√°les deber√≠an ser esos grupos.

* **¬øPara qu√© sirve?**

  * Segmentaci√≥n de clientes (marketing, bancos, e-commerce).
  * Detecci√≥n de anomal√≠as (fraude bancario, fallas en sensores).
  * Biolog√≠a (agrupar genes o c√©lulas con funciones similares).
  * Sistemas de recomendaci√≥n (agrupar usuarios con gustos parecidos).

* **Principales algoritmos:**

  * **K-Means** ‚Üí Divide los datos en *k* grupos minimizando la distancia al centroide. Es r√°pido pero sensible a valores at√≠picos.
  * **DBSCAN** ‚Üí Detecta clusters de distintas formas/tama√±os, adem√°s detecta *outliers*. Muy usado en datos geoespaciales.
  * **Clustering Jer√°rquico (HCA)** ‚Üí Construye una jerarqu√≠a de grupos (tipo √°rbol). Bueno para an√°lisis exploratorio.
  * **SVM de una clase (One-Class SVM)** ‚Üí Especializado en detectar si un dato es ‚Äúnormal‚Äù o una anomal√≠a.
  * **Isolation Forest** ‚Üí A√≠sla puntos an√≥malos en datasets grandes (muy usado en ciberseguridad y fraude).

---

### 2. **Reglas de Asociaci√≥n**

Aqu√≠ el objetivo es **encontrar relaciones frecuentes entre √≠tems o eventos** en un conjunto de datos.

* **¬øPara qu√© sirve?**

  * *Market Basket Analysis* (ejemplo cl√°sico: si compras pan, tambi√©n compras mantequilla).
  * Recomendadores de productos (Amazon, Mercado Libre).
  * Detecci√≥n de patrones en bioinform√°tica, redes sociales, seguridad.

* **Principales algoritmos:**

  * **Apriori** ‚Üí Encuentra reglas frecuentes combinando √≠tems paso a paso. F√°cil de entender, pero pesado con muchos datos.
  * **FP-Growth (Frequent Pattern Growth)** ‚Üí Optimizaci√≥n de Apriori que construye una estructura tipo √°rbol para mejorar velocidad.
  * **Eclat** ‚Üí Usa intersecci√≥n de conjuntos en vez de combinaciones, suele ser m√°s r√°pido en algunos escenarios.

---

### 3. **Reducci√≥n de Dimensionalidad**

Cuando los datos tienen much√≠simas variables (im√°genes, texto, gen√≥mica, big data), se busca **reducir dimensiones** pero conservando la mayor informaci√≥n posible.

* **¬øPara qu√© sirve?**

  * Preprocesar datos antes de aplicar clustering o clasificaci√≥n.
  * Visualizaci√≥n de datos en 2D/3D.
  * Compresi√≥n de im√°genes y reducci√≥n de ruido.
  * Evitar el problema de la ‚Äúmaldici√≥n de la dimensionalidad‚Äù (cuando demasiadas variables confunden al modelo).

* **Principales algoritmos:**

  * **PCA (Principal Component Analysis)** ‚Üí Proyecta los datos en nuevas variables (componentes principales) que maximizan la varianza.
  * **Kernel PCA** ‚Üí Variante de PCA que usa kernels para capturar relaciones no lineales.
  * **t-SNE (t-Distributed Stochastic Neighbor Embedding)** ‚Üí Excelente para visualizaci√≥n en 2D/3D de datos muy complejos. Muy usado en biolog√≠a computacional y NLP.
  * **LLE (Locally Linear Embedding)** ‚Üí Preserva la estructura local de los datos, √∫til para datos no lineales.

---

En resumen:

* **Clustering** = agrupar sin etiquetas.
* **Reglas de Asociaci√≥n** = descubrir patrones de co-ocurrencia.
* **Reducci√≥n de Dimensionalidad** = simplificar datos sin perder demasiada informaci√≥n.

# 7. **ALGORITMOS DE APRENDIZAJE AUTOM√ÅTICO**



**formas de entrenar y aplicar algoritmos de aprendizaje autom√°tico**.

---

## 1. **Aprendizaje por Lotes (Batch Learning)**

* El modelo se entrena con **todo el conjunto de datos de una sola vez**.
* Cada vez que llegan nuevos datos, hay que **volver a entrenar desde cero** con el dataset completo.

> **Ejemplo:** entrenar un modelo de predicci√≥n de fraudes con todos los registros de transacciones bancarias hist√≥ricas. Si entran nuevos clientes, se debe reentrenar el modelo completo.

‚úÖ Ventajas: mayor precisi√≥n en modelos grandes.

‚ùå Desventajas: consume muchos recursos y no se adapta r√°pido a cambios.

---

## 2. **Aprendizaje en L√≠nea (Online Learning)**

* El modelo **aprende de manera incremental**: se actualiza cada vez que llegan nuevos datos, sin necesidad de reentrenar todo.
* Ideal para sistemas donde los datos cambian r√°pido o llegan en flujo continuo (*streaming*).

> **Ejemplo:** un sistema de recomendaciones de YouTube/Netflix que va aprendiendo de lo que ves en tiempo real y adapta las sugerencias.

‚úÖ Ventajas: se adapta r√°pido a nuevos datos, eficiente en memoria.

‚ùå Desventajas: puede ser inestable si los datos que llegan son muy ruidosos.

---

## 3. **Aprendizaje Basado en Instancias (Instance-Based Learning)**

* El modelo **no construye reglas generales al entrenar**, sino que **guarda ejemplos del dataset** y decide comparando con ellos.
* Aprende ‚Äúmemorizar y comparar‚Äù m√°s que ‚Äúgeneralizar‚Äù.

> **Ejemplo:** el algoritmo **K-Nearest Neighbors (KNN)**. Para clasificar una nueva fruta, compara sus caracter√≠sticas (peso, color, tama√±o) con las frutas almacenadas y decide seg√∫n las m√°s cercanas.

‚úÖ Ventajas: simple y preciso si los datos est√°n bien representados.
‚ùå Desventajas: lento con grandes vol√∫menes de datos y sensible a ruido.

---

## 4. **Aprendizaje Basado en Modelos (Model-Based Learning)**

* El modelo **aprende una funci√≥n matem√°tica general** a partir de los datos de entrenamiento.
